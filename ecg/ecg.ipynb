```python
import matplotlib.pyplot as plt
%matplotlib notebook

import wfdb
import numpy as np
import scipy.signal

from IPython.display import display
```

```python
signals, fields = wfdb.rdsamp('data/mitdb/100')

signal = signals[:,0]

signal = scipy.signal.decimate(signal,4,ftype='fir')

d_signal = np.diff(signal)

#signals, fields
```

```python
plt.plot(signal)
plt.plot(d_signal)
```
